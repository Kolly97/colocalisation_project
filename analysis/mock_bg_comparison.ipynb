# Comparisons between Thomas and my results

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Output_dir = "/Volumes/Kolly_Stick/Universitaet/Bartenschlager_Internship/compare_mock_results/Huh7_rep1/"
```

```python
def column_plot_mock (
    labels = ["NS4B568", "HA488"],
    time = "24 h", 
    cell_type = "Huh7",
    ch1 = None,
    ch2 = None,
    Thomas_ctl = True,
    save = False,
    filename = None
):
    if filename is None:
        filename = f"Mock_threshold_comparison_{cell_type}_{time}_{labels[0]}_{labels[1]}.png"
    
    # Labels innerhalb jeder Gruppe
    if Thomas_ctl:
        sub_labels = ["Thomas", "p95", "p99", "p99.25", "p99.5", "p99.9", "p99.95", "p99.99", "p99.995", "p99.999"]
    else:
        sub_labels = ["p95", "p99", "p99.25", "p99.5", "p99.9", "p99.95", "p99.99", "p99.995", "p99.999"]
    
    
    # Anzahl Balken pro Gruppe
    n = len(sub_labels)
    
    # x Positionen
    x1 = np.arange(n)
    gap = 2
    x2 = np.arange(n) + n + gap
    
    # Figure
    fig, ax = plt.subplots(figsize=(14,6))
    
    # Balken
    ax.bar(x1, ch1, width=0.8, label=labels[0], edgecolor = "k")
    ax.bar(x2, ch2, width=0.8, label=labels[1], edgecolor = "k")

    if Thomas_ctl:
        # HA488 Thomas line
        ax.hlines(
            y=ch1[0],
            xmin=x1[0]-0.4,
            xmax=x1[-1]+0.4,
            colors="black",
            linestyles="dashed",
            linewidth=1.5,
            alpha=0.6
        )
        
        # NS4B568 Thomas line
        ax.hlines(
            y=ch2[0],
            xmin=x2[0]-0.4,
            xmax=x2[-1]+0.4,
            colors="black",
            linestyles="dashed",
            linewidth=1.5,
            alpha=0.6
        )
    
    # Einzelne Labels unter jedem Balken
    all_x = np.concatenate([x1, x2])
    all_labels = sub_labels + sub_labels
    
    ax.set_xticks(all_x)
    ax.set_xticklabels(all_labels, rotation=45)
    
    # Gruppentitel als zweite x-Achse
    secax = ax.secondary_xaxis('bottom')
    secax.set_xticks([np.mean(x1), np.mean(x2)])
    secax.set_xticklabels(labels, fontsize=14)
    secax.spines['bottom'].set_visible(False)
    secax.tick_params(axis='x', length=0, pad=45)
    
    # Layout
    ax.set_ylabel("Intensity value")
    #ax.set_yscale ("log")
    ax.set_title(f"Background comparison per channel \n {cell_type} {time} {labels[0]}_{labels[1]} combo ", fontsize = 14, fontweight = "bold")
    
    plt.tight_layout()
    plt.subplots_adjust(bottom=0.25)

    if save:
        plt.savefig(
            f"{Output_dir}{filename}",
            dpi=600,
            bbox_inches="tight"
        )
    
    plt.show()
```
